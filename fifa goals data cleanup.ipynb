Clean up FIFA Goals data
source file: data cleaning\fifa goals\fifa-worldcup-goals-cleanup.xlsx

```python
import pandas as pd

df = pd.read_excel(r"fifa-worldcup-goals-cleanup.xlsx")
df
```

Lets combine all goals to a 3 column table - date, country, goal detail

```python
df_home = df[['Date','Home Goals','Home Country']]
df_away = df[['Date','Away Goals','Away Country']]
df_home.columns = ['Date', 'Goal Detail','Country']
df_away.columns = df_home.columns
df_all  = pd.concat([df_home, df_away])
df_all = df_all.dropna()
df_all
```

Let's split the goals by new line

```python
df_all['Goals'] = df_all['Goal Detail'].str.split('\n')
df_all = df_all.explode('Goals')
df_all
```

Let's split the goal times and explode them as well

```python
df_all[['Player', 'Minutes Raw']] = df_all['Goals'].str.split(r"(?=\d)", n=1, expand=True)
df_all['Minute'] = df_all['Minutes Raw'].str.split(",")
df_all = df_all.explode('Minute').reset_index(drop=True)
df_all["Player"] = df_all["Player"].str.strip()
df_all["Minute"] = df_all["Minute"].str.strip()
df_all
```

Let's remove the columns we no longer need and 
clean up OG & Pen stuff and add them as a column to tag the goal

```python
import numpy as np

goal_checks = [
    df_all['Minute'].str.contains('(o.g.)',regex=False, na=False),
    df_all['Minute'].str.contains('(pen.)', regex=False, na=False),
]
goal_types = ['Own Goal', 'Penalty']

df_all = df_all[['Date','Country','Player','Minute']].copy()
df_all['Goal Type'] = np.select(goal_checks, goal_types, default="Regular")
df_all
```
